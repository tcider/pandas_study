    + 3) Плата за получение наличных. Россия

```python
### Type your code here
tr = transactions[transactions.amount > 0]
tr = tr[['tr_type', 'gender', 'amount']]
tr = tr.dropna()
tr0 = tr[tr.gender == 0]
trg0 = tr0.groupby(['tr_type', 'gender']).agg([('max_income', 'max')]).amount.sort_values(by = 'max_income')[0:5]
tr1 = tr[tr.gender == 1]
trg1 = tr1.groupby(['tr_type', 'gender']).agg([('max_income', 'max')]).amount.sort_values(by = 'max_income')[0:5]
tr_text = transactions[['tr_description', 'tr_type']].drop_duplicates()
tr_new = pd.concat([trg0, trg1]).reset_index(['gender'])
pd.merge(tr_text, tr_new, on = 'tr_type', how='right').sort_values(by = 'gender')
```

##### Пояснения:

функцией .apply(lambda x: x[])

```python
def plot_pivot_table(pivot_table):
    plt.figure(figsize=(9, 11))
    sns.heatmap(pivot_table, cmap="YlGnBu", annot=True, 
                fmt='.3g', annot_kws={"size": 14, "fontsize": 14})
    plt.xticks(fontsize=15)
    plt.yticks(rotation=0, fontsize=15)
    plt.xlabel('Bucket', size=18)
    plt.ylabel('Hour', fontsize=18)
    plt.title('Gender analysis per bucket and hour', fontsize=20)
    plt.show()
```

```python
# Type your code here
pd.set_option('display.max_rows', 100)
#transactions['amount_bucket'] = pd.qcut(transactions[transactions.amount < 0].amount, 5, labels = ['Very High', 'High', 'Middle', 'Low', 'Very Low'])
#transactions['amount_bucket'].cat.add_categories('Income')
#transactions.amount_bucket.fillna('Income')
#transactions
amount_bucket = pd.qcut(transactions[transactions.amount < 0].amount, 5, labels = ['Very High', 'High', 'Middle', 'Low', 'Very Low'])
amount_bucket.cat.add_categories('Income')
amount_bucket2 = pd.qcut(transactions[transactions.amount > 0].amount, 1, labels = ['Income'])
amount_bucket = pd.concat([amount_bucket, amount_bucket2])
transactions['amount_bucket'] = amount_bucket
transactions['tr_hour'] = transactions['tr_datetime'].map(lambda s: (s.split(' ')[1]).split(':')[0])
res = transactions.pivot_table('gender', index='tr_hour', columns='amount_bucket')
res
plot_pivot_table(res)
```
